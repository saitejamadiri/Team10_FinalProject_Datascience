```python
train = df.loc[:, ["Open"]].values
train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1)) # Converts between 0 and 1
train_scaled = scaler.fit_transform(train)
train_scaled
```

```python
plt.plot(train_scaled)
plt.ylabel("Price")
plt.xlabel("Time(Days)")
plt.title("Google Stock Data")
plt.show()
```

```python
# Creating a data structure with 50 timesteps and 1 output
X_train = []
y_train = []
timesteps = 1
```

```python
for i in range(timesteps, 1258): #1258: len of days
    X_train.append(train_scaled[i-timesteps:i, 0])
    y_train.append(train_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_train
```

```python
y_train
```

```python
print(f"Shape of X_train: {X_train.shape}\nShape of y_train: {y_train.shape}")
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
pip install numpy==<version>
```

```python
model = Sequential()
model.add(LSTM(units = 50,return_sequences = True,input_shape = (X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units = 50,return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(units = 50,return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(units = 50))
model.add(Dropout(0.2))
model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam',loss = 'mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train.reshape(-1,1), epochs=100)
```

```python
import csv

def array_to_csv(data, file_path):
    """
    Converts an array to a CSV file.

    :param data: The array to convert to CSV.
    :param file_path: The file path where the CSV file will be saved.
    """
    with open(file_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        for row in data:
            writer.writerow(row)

# Example array
# data = [
#     ['Name', 'Age', 'Country'],
#     ['John', 30, 'USA'],
#     ['Emily', 25, 'Canada'],
#     ['David', 40, 'UK']
# ]

# File path to save CSV
file_path = 'output.csv'

# Convert array to CSV and save to file
array_to_csv(inputs, file_path)
```

```python
import boto3

def upload_to_s3(file_path, bucket_name, object_name=None):
    """
    Uploads a file to an S3 bucket.

    :param file_path: Path to the file to upload.
    :param bucket_name: Name of the S3 bucket.
    :param object_name: S3 object name. If not specified, file_name is used.
    :return: True if file was uploaded, else False.
    """
    # If S3 object_name was not specified, use file_name
    if object_name is None:
        object_name = file_path

    # Create an S3 client
    s3_client = boto3.client('s3')

    try:
        # Upload the file
        response = s3_client.upload_file(file_path, bucket_name, object_name)
    except Exception as e:
        print(e)
        return False
    return True

# Example usage
file_path = 'output.csv'  # Path to the file you want to upload
bucket_name = 'stockpriceprediction-01'  # Name of your S3 bucket

# Upload the file
upload_to_s3(file_path, bucket_name)
```
